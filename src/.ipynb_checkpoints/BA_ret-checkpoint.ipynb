```python
import pandas as pd
import numpy as np
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12.0, 8.0)
plt.rcParams.update({'font.size': 18})
print(plt.style.available)
```

```python
ret = pd.DataFrame()
for epsilon in np.linspace(0.001, 0.01, 10):
    if epsilon == 0.001:
        data = pd.read_csv('bilevel_opt_epsilon=%.5f_attackBestResponse_BA.txt' % epsilon, header=None, sep=' ')
        data = data.iloc[:, [1,3]]
    else:
        data = pd.read_csv('bilevel_opt_epsilon=%.5f_attackBestResponse_BA.txt' % epsilon, header=None, sep=' ')
        data = data.iloc[:, 3]
    ret = pd.concat([ret, data], axis=1)
ret.columns = ['algo'] + ['%.3f' % item for item in np.linspace(0.001, 0.01, 10)]
```

```python
ret.set_index('algo').transpose().plot(kind='bar')
plt.xticks(rotation=45)
plt.xlabel('budget')
plt.ylabel('$U_d$')
plt.title('Barabasi Albert')
```
